```python
import pandas as pd #used to read txt file
import numpy as np #used to handle arrays
import matplotlib.pyplot as plt #used to plot data
import seaborn as sns
```

```python
def plotData(dataset, xColumn, yColumn, xLabel, yLabel):
    for marker in ["x", "o", "*"]:
        for color in ["r", "g", "b"]:
            plt.figure(figsize=(20,10))
            plt.scatter(x=xColumn, y=yColumn, data=dataset, c=color, marker=marker)
            plt.xlabel(xLabel)
            plt.ylabel(yLabel)
            plt.legend()
            plt.show()
```

```python
def computeCost(X, y, theta):
    cost = 0
    for i in range(len(y)):
        hypothesis = theta[0] + theta[1] * X[1][i]
        cost += (hypothesis - y[i])**2
    cost /= 2 * len(y)
    return cost
```

```python
def gradientDescent(X, y, theta, alpha, num_iters, lambdaParameter=1):
    if(lambdaParameter == 1):
        lambdaParameter = alpha / len(y)
        
    for i in range(num_iters):
        
        temp1 = 0
        temp2 = 0

        for obsNumber in range(len(y)):
            hypothesis = theta[0] + theta[1] * X[1][obsNumber]
            temp1 += hypothesis - y[obsNumber]
            temp2 += (hypothesis - y[obsNumber]) * X[1][obsNumber]
            
        temp1 *= alpha / len(y)
        temp2 *= alpha / len(y)
        
        theta[0] -= temp1
        theta[1] = (theta[1] * (alpha * lambdaParameter / len(y))) - (temp2)
        
        if(i % 100 == 0):
            print("Step {}: theta {}".format(i, theta))
        
        
```

```python
def testGradientDescentLinearRegression(X, y, iterations, alpha):
    theta = np.zeros((2,1))
    
    X = np.array((np.zeros_like(X), X))
    
    initialCost = computeCost(X, y, theta)
    print("Initial cost: ", initialCost)
    gradientDescent(X, y, theta, alpha, iterations)
    
    print("Optimal Theta: ", theta)
    return theta
```

```python
def computeMSE(X, y, theta):
    squared_errors = np.zeros_like(y)
    predictions = np.zeros((len(y)))
    mse = 0
    for i in range(len(squared_errors)):
        hypothesis = theta[0] + theta[1] * X[i]
        predictions[i] = hypothesis
        squared_error = (y[i] - hypothesis)**2
        squared_errors[i] = squared_error
        mse += squared_error
    mse /= len(y)
    print("Mean Squared Error : ", mse)
    print("Root Mean Squared Error: ", mse**(0.5))
    plt.figure(figsize=(20,10))
    plt.scatter(X, squared_errors, label="Squared Error")
    plt.legend()
    plt.show()

    plt.figure(figsize=(20,10))
    plt.scatter(x=X, y=y, c="b", marker="x")
    plt.plot(X, predictions, "r", label="Linear model")
    plt.legend()
    plt.show()
```
